Question 1: 
Write a program which will find all such numbers which are divisible by 7 but are not a multiple of 5, between 2000 and 3200 (both included).The numbers obtained should be printed in a comma-separated sequence on a single line.

Hints: 
Consider use range(#begin, #end) method

```python
l=[]
for i in range(2000, 3201):
    if (i%7==0) and (i%5!=0):
        l.append(str(i))

print(','.join(l))
```

UNDERSTANDING & PRACTICE

#IN THE RANGE 2000 TO 3200 (BOTH NUMBERS INCLUDED), FIND NUMBERS THAT ARE DIVISIBLE BY 7 BUT NOT BY 5. DISPLAY THESE NUMBERS IN A SINGLE LINE SEPARATED BY COMMA.

```python
#range(2000, 3201) #This range includes both 2000 and 3200
#for n in range (2000,3201): #For loop will be used to iterate one number at a time in given range
    #if(n%7==0) and (n%5!=0): #This is the actual logic for checking that the number is divisible by 7 and not by 5. The modulo (%) returns remainder after dividing number by 7 and by 5. Equals is == and Not Equals is !=
#list[] #A list is required for the result of logic of each iteration to be saved/appended into its individual index positions
    #list.append(str(n)) #within the if condition, the result of each iteration is to be saved into pre-defined empty list (which we named list). We are doing this by appending the result number (which we named n in for loop) to the corresponding index in sequence within the pre defined list. Only strings can be appended so we are converting integer or float data to string type.
#print(",".join(list)) #we are now displaying the resultant list elements in a comma separated line using .join ***

list=[]
for n in range(2000, 3201):
    if(n%7==0) and (n%5!=0):
        list.append(str(n))
print(",".join(list))
```
